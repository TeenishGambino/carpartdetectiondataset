```python
import numpy as np
import os

from tflite_model_maker.config import ExportFormat
from tflite_model_maker import model_spec
from tflite_model_maker import object_detector

import tensorflow as tf
assert tf.__version__.startswith('2')

tf.get_logger().setLevel('ERROR')
from absl import logging
logging.set_verbosity(logging.ERROR)
```

```python
spec = model_spec.get('efficientdet_lite2')
```

```python
train_data = object_detector.DataLoader.from_pascal_voc('/tf/CARS.v1-initial-dataset.voc/train', '/tf/CARS.v1-initial-dataset.voc/train', 
["back_bumper",
 "back_glass",
"back_left_door",
"back_left_light",
"back_right_door",
"back_right_light",
"front_bumper",
"front_glass",
"front_left_door",
"front_left_light",
"front_right_door",
"front_right_light",
"hood",
"left_mirror",
"right_mirror",
"tailgate",
"trunk",
"wheel"])
```

```python
validation_data = object_detector.DataLoader.from_pascal_voc('/tf/CARS.v1-initial-dataset.voc/valid', '/tf/CARS.v1-initial-dataset.voc/valid', 
["back_bumper",
 "back_glass",
"back_left_door",
"back_left_light",
"back_right_door",
"back_right_light",
"front_bumper",
"front_glass",
"front_left_door",
"front_left_light",
"front_right_door",
"front_right_light",
"hood",
"left_mirror",
"right_mirror",
"tailgate",
"trunk",
"wheel"])
```

```python
model = object_detector.create(train_data, model_spec=spec, epochs=100, batch_size=4, train_whole_model=True, validation_data=validation_data)
```

```python
model.evaluate(validation_data)
```

```python
model.export(export_dir='/tf/')
```
